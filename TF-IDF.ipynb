```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

corpus = ["O cachorro gosta de correr",
          "O gato gosta de dormir",
          "O peixe gosta de nadar",
          "O tubarão adora sangue"]

# Bag of Words
vectorizer = CountVectorizer()
X_bow = vectorizer.fit_transform(corpus).toarray()
print("BoW:\n", X_bow)

# TF-IDF
tfidf = TfidfVectorizer()
X_tfidf = tfidf.fit_transform(corpus).toarray()
print("TF-IDF:\n", X_tfidf)
```

```python
'''
Bag of Words (BoW) - Conceito
➔ O que é?
Uma maneira simples e poderosa de representar um texto como vetores de frequência de palavras.

---------- ---------- ----------

TF-IDF (Term Frequency - Inverse Document Frequency)
➔ O que é?
Uma evolução do BoW que dá peso às palavras mais importantes e reduz o peso das palavras comuns.
'''
```
